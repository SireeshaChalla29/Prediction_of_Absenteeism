# Importing

```python
#importing all the necessary functions from their respective libraries
import pandas as pd
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
from PIL import Image

from sklearn import tree
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
data= pd.read_csv(r"C:\Users\Dell\Downloads\Absenteeism_at_work.csv") #loading the dataset 
data.head()
```

```python
#Extracting the target variable "absenteeism in hours" from the data set 

y = data['Absenteeism time in hours']
data.drop('Absenteeism time in hours', axis=1, inplace=True)
```

```python
# Split data into training  and validation data to accurately predict 

X_train, X_valid, y_train, y_valid = train_test_split(data, y, test_size=0.2, shuffle=True)
```

```python
#using DecisionTreeRegressor function to build regression model in the form of tree structure

model = DecisionTreeRegressor(max_depth=6, min_samples_leaf=26, random_state=42) 
```

```python
#Fit function adjusts weights according to data values so that better accuracy can be achieved
model.fit(X_train, y_train)
```

```python
def score(y_actual, y_preds):
   return sqrt(mean_squared_error(y_actual, y_preds)) 
```

```python
score(y_train, model.predict(X_train)) #score of train data set 
```

```python
score(y_valid, model.predict(X_valid)) #score of validation data set 
```

```python
#finding the mean squared error of validation data set  using mean_squared_error function

mean_squared_error(y_valid, model.predict(X_valid)) 
```

```python
#finding the mean squared error of training data set  using mean_squared_error function

mean_squared_error(y_train, model.predict(X_train))
```

```python
#finding the  mean absolute error of validation data set  using mean absolute error function 

mean_absolute_error(y_valid, model.predict(X_valid))
```

```python
#finding the  mean absolute error of training data set  using mean absolute error function 
mean_absolute_error(y_train, model.predict(X_train))
```

# Plotting the tree

```python
#plotting the tree using the plot function 
model = DecisionTreeRegressor(max_depth=6, min_samples_leaf=26, random_state=42)
plt.figure(figsize=(17,17))
tree.plot_tree(model.fit(X_train, y_train));
plt.show()
```
